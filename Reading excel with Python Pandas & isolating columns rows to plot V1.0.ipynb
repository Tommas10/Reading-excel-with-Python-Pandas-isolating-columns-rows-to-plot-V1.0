```python
#將在大型數據框的第31行中繪製數據，並將第0行設置為x軸.
#Created by Tommas Huang Date: 2019-11-30
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#創建一個具有32行10列的隨機數組
df = pd.DataFrame(np.random.rand(320).reshape(32,10), columns=range(64,74), index=range(1,33))
df.to_excel("/Users/TommasHuang/Downloads/TimeSeriesPythonPandaPlot/QuantaTrainingStocks2019.xlsx")
```

```python
#使用“ parse_cols”和“ skiprows”僅讀取所需的列和行。 本例子中的第一列是數據 frame索引。

# load desired columns and rows into a dataframe
# in this method, I firse make a list of all skipped_rows
desired_cols = [0] + list(range(2,9))
skipped_rows = list(range(1,33))
skipped_rows.remove(31)
df = pd.read_excel("/Users/TommasHuang/Downloads/TimeSeriesPythonPandaPlot/QuantaTrainingStocks2019.xlsx", index_col=0, usecols=desired_cols, skiprows=skipped_rows)
```

```python
#僅隔離要繪製的行，以原始列標題為索引的pandas.Series
ser = df.loc[31, :]
```

```python
#繪製系列。
fig, ax = plt.subplots()
ser.plot(ax=ax)
ax.set_xlabel("year")
ax.set_ylabel("precipitation")
```

```python
fig, ax = plt.subplots()
ser.plot(kind="bar", ax=ax)
ax.set_xlabel("year")
ax.set_ylabel("precipitation")
```
